# Liner Algebra 


```python
import numpy as np
```

```python
x = np.array([1,2,3])
y = 2 * x

n = len(x)
w = 0
lr = 0.01
for _ in range(100):
    Y = w * x
    dw = (2/n) * np.sum((Y-y) * x)

    w -= lr * dw

print(w)
```

# Gradient and Optimisation (Cost Function)

```python
import sympy as sp
```

```python
w = sp.symbols("w")
J = (w-5)**2
f_diff = sp.diff(J, w)
print(f_diff)
```

```python
x = np.array([1,2,3,4,5])
y = (w-5)**2

n = len(x)
w = 0
lr = 0.1

for _ in range(5):
    Y = (w-5)**2
    dw = (2/n)*(np.sum(Y-y))

    w -= lr*dw
    print(w)
```
