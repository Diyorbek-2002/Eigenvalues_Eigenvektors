```python
import numpy as np
def binary_search(arr, item):
    low = 0 
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == item:
            return mid
        elif arr[mid] < item:
            low = arr[mid] + 1
        else:
            high = arr[mid] - 1
    return None

print(binary_search(np.array([1,2,3,4,5,6,7,8,8,9]), 5))
            
            
```

```python
def selection_sort(arr):
    new_arr = []
    for _ in range(len(arr)):
        smallest = min(arr)
        new_arr.append(smallest)
        arr.remove(smallest)
    return new_arr
print(selection_sort([3,2,1,4,5,6,74,3,2]))
```
